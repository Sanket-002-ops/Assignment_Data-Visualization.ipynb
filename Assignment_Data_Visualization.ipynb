# **Question5: Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.**

```python
# Generate synthetic dataset with correlated features
feature_1 = np.random.rand(100) * 100
feature_2 = 2 * feature_1 + np.random.normal(0, 10, 100)
feature_3 = -0.5 * feature_1 + np.random.normal(0, 5, 100)
feature_4 = feature_2 + feature_3 + np.random.normal(0, 5, 100)

data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Compute correlation matrix
corr_matrix = data.corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Matrix Heatmap of Synthetic Dataset")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed
np.random.seed(25)

# Create dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.6, edgecolors="w", linewidth=2)

# Add labels
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], country, fontsize=14, ha='center')

# Labels and title
plt.title('Bubble Chart of Population vs GDP by Country', fontsize=19)
plt.xlabel('GDP (in billions)', fontsize=14)
plt.ylabel('Population (in millions)', fontsize=14)

# Show plot
plt.show()
```

# **Question1: Create a Bokeh plot displaying a sine wave. Set x-value from 0 to 10 and y-values as the sine of x.**

```python
from bokeh.plotting import figure, output_notebook, show
import numpy as np

# Enable output in the notebook
output_notebook()

# Prepare data
x = np.linspace(0, 10, 100)  # x-values from 0 to 10
y = np.sin(x)                # y-values as sine of x

# Create a new plot
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)", width=700, height=400)

# Add line to the plot
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

# **Question2: Create a Bokeh plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.**

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np
import random

# Generate random data
np.random.seed(42)  # For reproducibility
x = np.random.rand(100) * 10  # Random x values
y = np.random.rand(100) * 10  # Random y values
sizes = np.random.randint(10, 30, size=100)  # Random sizes for markers
colors = [f"#{random.randint(0, 0xFFFFFF):06x}" for _ in range(100)]  # Random hex colors

# Create ColumnDataSource to hold data
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a scatter plot
p = figure(title="Random Scatter Plot", x_axis_label="X", y_axis_label="Y", width=700, height=400)
p.scatter("x", "y", size="sizes", color="colors", alpha=0.6, source=source)

# Show the plot
show(p)
```

# **Question3: Generate a Bokeh bar chart representing the counts of different fruites using the following dataset.**

```python
#fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
#counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource from the dataset
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruit", y_axis_label="Count", width=700, height=400)
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="black")

# Show the plot
show(p)
```

# **Question5: Create a Bokeh heatmap using the provided dataste.**

```python
#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.layouts import layout
from bokeh import palettes
import numpy as np
import pandas as pd

# Sample data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh
df = pd.DataFrame({
    'x': xx.ravel(),
    'y': yy.ravel(),
    'value': data_heatmap.ravel()
})

# Create color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create figure
p = figure(title="Heatmap Example", tools="hover", tooltips=[("Value", "@value")],
           x_axis_label='X', y_axis_label='Y', x_range=(0, 1), y_range=(0, 1))

# Add rect glyph for heatmap
p.rect(x="x", y="y", width=0.1, height=0.1, source=df,
       fill_color=transform('value', color_mapper), line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Display plot
output_notebook()
show(p)
```
